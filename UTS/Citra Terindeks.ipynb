```python
from PIL import Image
import math
import matplotlib.pyplot as plt
```

```python
# Definisikan palet warna
palette = {
    "Hitam": (0, 0, 0),
    "Merah": (255, 0, 0),
    "Hijau": (0, 255, 0),
    "Biru": (0, 0, 255)
}
```

```python
# Fungsi jarak Euclidean
def euclidean_distance(rgb1, rgb2):
    return math.sqrt(
        (rgb1[0] - rgb2[0])**2 +
        (rgb1[1] - rgb2[1])**2 +
        (rgb1[2] - rgb2[2])**2
    )
```

```python
# Buka gambar
img = Image.open('pemandangan.jpg')  # Ganti dengan file gambar kamu
img = img.convert('RGB')  # Pastikan RGB
width, height = img.size
```

```python
plt.imshow(img)
plt.axis('off')  # Hilangkan sumbu
plt.title('Gambar Asli')
plt.show()
```

```python
# Matriks hasil
result_matrix = []

for y in range(height):
    row = []
    for x in range(width):
        pixel = img.getpixel((x, y))
        
        # Cari warna terdekat
        min_distance = float('inf')
        closest_color = (0, 0, 0)
        
        for color_name, color_rgb in palette.items():
            distance = euclidean_distance(pixel, color_rgb)
            if distance < min_distance:
                min_distance = distance
                closest_color = color_rgb
        
        row.append(closest_color)
    result_matrix.append(row)
```

```python
result_matrix
```

```python
# Buat gambar baru dari hasil
new_img = Image.new('RGB', (width, height))

for y in range(height):
    for x in range(width):
        new_img.putpixel((x, y), result_matrix[y][x])
```

```python
# Tampilkan gambar hasil menggunakan matplotlib
plt.imshow(new_img)
plt.axis('off')  # Hilangkan sumbu
plt.title('Hasil Konversi Warna')
plt.show()
```
