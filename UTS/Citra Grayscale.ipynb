```python
from PIL import Image
import math
import matplotlib.pyplot as plt
```

```python
def convert_to_grayscale(pixel):
    r, g, b = pixel

    gray = int(0.299 * r + 0.587 * g + 0.114 * b)

    return gray
```

```python
# Buka gambar
img = Image.open('pemandangan.jpg') 
img = img.convert('RGB')
width, height = img.size
```

```python
plt.imshow(img)
plt.axis('off')
plt.title('Gambar Asli')
plt.show()
```

```python
result_matrix = []
```

```python
for y in range(height):
    row = []
    for x in range(width):
        pixel = img.getpixel((x, y))
        index = convert_to_grayscale(pixel)
        row.append(index)
    result_matrix.append(row)
```

```python
# Tampilkan gambar hasil menggunakan matplotlib
plt.imshow(result_matrix, cmap='gray')
plt.axis('off')  # Hilangkan sumbu
plt.title('Hasil Konversi Warna')
plt.show()
```
